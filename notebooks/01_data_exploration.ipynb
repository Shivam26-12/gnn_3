## 1.3 Sales Statistics

```python
sales = dfs['sales']
day_cols = [c for c in sales.columns if c.startswith('d_')]
sales_matrix = sales[day_cols].values

print(f'Shape: {sales_matrix.shape}')
print(f'Total sales: {sales_matrix.sum():,.0f}')
print(f'Mean daily sales per item: {sales_matrix.mean():.3f}')
print(f'% zeros (potential stockouts): {(sales_matrix == 0).mean() * 100:.1f}%')
print(f'Max single-day sale: {sales_matrix.max()}')
```

## 1.4 Calendar & Events

```python
cal = dfs['calendar']
print(f"Events: {cal['event_name_1'].dropna().nunique()} unique")
print(f"SNAP days: CA={cal['snap_CA'].sum()}, TX={cal['snap_TX'].sum()}, WI={cal['snap_WI'].sum()}")
print(f"\nEvent types:")
print(cal['event_type_1'].value_counts())
```
